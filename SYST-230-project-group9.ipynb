```python
#run second

def sup():
    t.tracer(0)

def draw(): #render screen
    sup()
    t.clearscreen() #new frame
    sup()
    t.hideturtle()
    sup()
    draw_port() #draw airport
    draw_runway(runways) #draw(n) n:number of runways
    sim_planes() #draw planes and move them
    t.update() #draw to screen

def draw_port(): #draw rectangle
    t.pu() #do not draw
    t.goto(-50, -30) #move turtle here: (x,y)
    t.pd() #start to draw
    t.fd(100) #move foward by 100
    t.left(90) #turn left 90 degrees
    t.fd(60)
    t.left(90)
    t.fd(100)
    t.left(90)
    t.fd(60)
    t.left(90)
    t.pu() #do not draw


def draw_runway(j): #draw j number of runways
    t.goto(-50, -50)
    t.pd()
    t.forward(100)
    t.goto(-50, -50)
    t.pu()
    for j in range(runways): # draw j numbers of rectangles
        t.pd()
        t.right(90)
        t.fd(30)
        t.left(90)
        t.fd(100)
        t.left(90)
        t.fd(30)
        t.right(90)
        t.pu()
        t.goto(-50, -50 - 30*(j+1)) #next j rectangle start point


def draw_plane(x, y, f): #pos: (x, y) rotation: f       #draws triangle
    t.pu()
    t.goto(x,y) #triangle's center
    t.left(f)
    t.fd(10 * m.sqrt(3) / 3)
    t.pd()
    t.left(180-30)
    t.fd(10)
    t.left(120)
    t.fd(10)
    t.left(120)
    t.fd(10)
    t.right(f+30)
    t.pu


def sim_planes(): #draw and move planes #most important func, connects all of the other funcs
    for j in planes: #spawn in planes
        if j.h < 0:
            j.h = r.randrange(-298, 1, 1)
            if j.h >= 0: #spawn plane
                away_away(j) #rand tp
                j.h = 100

    for j in planes: #for each plane, move by step needed
        if j.h >= 0: #if spawned in
            #steps
            if j.gotob == True: #go to airport
                base_path(j)
            elif j.cirwait == True: #circle if runway full
                cir_base(j)
            elif j.landperms == True: #land
                landing(j)
            elif j.bord == True: #get passangers
                get_pogs(j)
            elif j.gotob == False: #take off
                go_away(j)
            elif j.air == True: #default move
                j.x += 10 * m.cos(m.pi * j.f / 180)
                j.y += 10 * m.sin(m.pi * j.f / 180)
            else: #do nothing if not in air
                pass
            draw_plane(j.x, j.y, j.f) #draw plane on screen

def base_path(j): #go to base
    if m.sqrt(j.x**2 + j.y**2) > rad: #far from bass
        if j.x >= 0:
            if 180 * m.atan(j.y / j.x) / m.pi + 180 - j.h > 10: #if angle is off
                j.f = 180 * m.atan(j.y / j.x) / m.pi + 180 #fix angle, point toward base
        else:
            if 180 * m.atan(j.y / j.x) / m.pi - j.h > 10: #if angle is off
                j.f = 180 * m.atan(j.y / j.x) / m.pi #fix angle, point toward base

        j.x += 10 * m.cos(m.pi * j.f / 180) #moves
        j.y += 10 * m.sin(m.pi * j.f / 180)
    else: #close enough to base
        l = 0
        for u in run.load: #number of open runways
            if u == 0:
                l += 1
        ind = 0
        for u in run.load: #first open runway
            if u == 0:
                break
            ind += 1
        if len(run.load) - l >= runways: #is open runway
            j.gotob = False
            j.cirwait = True #cirle_base() will activate next frame for j, this plane
            j.f -= 90
        else: #land at base in open runway
            j.gotob = False
            j.landperms = True
            j.cirwait == False
            run.load[ind] = j

def cir_base(j): #wait for open port
    if m.sqrt(j.x**2 + j.y**2) > rad: #move in close enough circle
        j.f += 7 #face closer to center
    j.x += 10 * m.cos(m.pi * j.f / 180) #move
    j.y += 10 * m.sin(m.pi * j.f / 180)

    l = 0
    for u in run.load: #number of open runways
        if u == 0:
            l += 1

    if len(run.load) - l >= runways:
        pass #keep circling if no open runway
    else: #land
        ind = 0 #find open runway
        for u in run.load:
            if u == 0:
                break
            ind += 1
        j.cirwait = False
        j.landperms = True
        run.load[ind] = j
        k = 0 #pick open runway
        for i in run.load:
            if i == j:
                break
            k += 1
        if j.x >= 0: #face open runway, depends on quadrant
            j.f = 180 * m.atan((-50 - 30 * (k + 1) - j.y) / -j.x) / m.pi + 180
        else:
            j.f = 180 * m.atan((-50 - 30 * (k + 1) - j.y) / -j.x) / m.pi

def landing(j): #land on open runway
    k = 0 #pick open runway
    for i in run.load:
        if i == j:
            break
        k += 1

    if j.x >= 0: #face picked runway
        if 180 * m.atan((-50 - 30 * (k + 1) - j.y) / (-j.x)) / m.pi + 180 - j.f > 10:
            j.f = 180 * m.atan((-50 - 30 * (k + 1) - j.y) / -j.x) / m.pi + 180
    else:
        if 180 * m.atan((-50 - 30 * (k + 1) - j.y) / (-j.x)) / m.pi - j.f > 10:
            j.f = 180 * m.atan((-50 - 30 * (k + 1) - j.y) / -j.x) / m.pi

    if m.sqrt((0-j.x)**2 + (-50 - 30 * (k + 1) - j.y)**2) > 15: #far from runway
        j.x += 10 * m.cos(m.pi * j.f / 180) #move
        j.y += 10 * m.sin(m.pi * j.f / 180)
    else: #close enogh to runway, landed, get_pogs()
        j.landperms = False
        j.bord = True
        j.gotopogs = True
        j.air = False
        j.f = 0
        run.enter += 1 #update for pandas

def get_pogs(j): #get passangers
    if j.gotopogs == True: #goto hanger

        j.x += 10 * m.cos(m.pi * j.f / 180)
        j.y += 10 * m.sin(m.pi * j.f / 180)
        if j.y < 0:
            if j.x > 70:
                j.f = 90
        elif j.x > 0:
            j.f = 180 #turn
        else:
            j.gotopogs = False
            k = 0 #get on runway pos
            for i in run.load:
                if i == j:
                    break
                k += 1
            run.load[k] = 0 #remove plane from runway
            run.hanger.append(j) #add plane to hanger
    elif j in run.hanger: #
        if j.landtime <= 0: #leave with pogs
            j.pogs = r.randrange(1, 20, 1)
            run.hanger.remove(j) #remove self from hanger
            run.qout.append(j) #add self to queue to leave
        else: #wait for pogs
            j.landtime -= 1
    elif j not in run.load: #in qout
        l = 0
        for u in run.load: #find open runways
            if u == 0:
                l += 1
        if len(run.load) - l >= runways: #no open runway
            if j.x > -70: #move in position to leave
                j.f = 180
                j.x += 10 * m.cos(m.pi * j.f / 180)
                j.y += 10 * m.sin(m.pi * j.f / 180)
        else:
            ind = 0
            for u in run.load: #pick runway
                if u == 0:
                    break
                ind += 1
            run.load[ind] = j #add to picked runway
            run.qout.remove(j) #remove self from queue to leave
    else: #take off
        k = 0 #find current runway
        for i in run.load:
            if i == j:
                break
            k += 1
        if m.sqrt((0-j.x)**2 + (-50 - 30 * (k + 1) - j.y)**2) < 11: #take off spot
            j.f = r.randrange(0, 360, 1)
            j.bord = False
            j.gotob = False
            j.air = True #take off
            run.load[k] = 0 #no longer on runway
            run.exit += 1 #update for pandas
        else: #goto take off spot
            if (-50 - 30 * (k + 1)) < j.y:
                j.f = 270
            elif j.x <= 0:
                j.f = 0
            j.x += 10 * m.cos(m.pi * j.f / 180)
            j.y += 10 * m.sin(m.pi * j.f / 180)


def go_away(j): #take off or going away
    if m.sqrt(j.x**2 + j.y**2) > 700: #700 away from center #off screen
        away_away(j) #tp to rand location
    else: #move
        j.x += 10 * m.cos(m.pi * j.f / 180)
        j.y += 10 * m.sin(m.pi * j.f / 180)

def away_away(j): #rand tp
    ang = r.randrange(0, 360, 1)
    leng = r.randrange(600, 800, 1)
    j.x = leng * m.cos(m.pi * ang / 180)
    j.y = leng * m.sin(m.pi * ang / 180)
    j.gotob = True
    j.pogs = r.randrange(1, 20, 1)
    j.landtime = r.randrange(20, 60, 1)


class plane: #pos (x,y), hieght: h, facing: f #in degrees
    def __init__(self, x, y, h, f):
        self.x = x
        self.y = y
        self.h = h
        self.f = f

    air = True #is in air
    cirwait = False #waiting to land
    landperms = False #ready to land
    pogs = 0 #passangers
    gotopogs = False #after landing
    bord = False #bording
    landtime = 20 #waittime to leave when at hanger
    gotob = False #gotobase: #True = needing to land, False = leaving

class runway: #number of runways: num
    def __init__(self, num):
        self.num = num

    load = [] #number of planes on the runway or going to land or leaving
    hanger = [] #number of planes in hanger and time when leave
    qout = [] #queue wanting to leave
    enter = 0
    exit = 0
```
